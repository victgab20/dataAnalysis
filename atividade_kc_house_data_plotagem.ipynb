```python
import pandas as pd
import plotly.express as px
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
arquivo = '/content/drive/MyDrive/Colab Notebooks/kc_house_data.csv'
dataset = pd.read_csv(arquivo, sep=',' ,header=0)
```

```python
dataset.info()
```

```python
print("Vamos fazer um levamento da faixa de preço")
valor_maximo = int(input("informe o valor máximo que você está disposta a pagar"))
valor_minimo = int(input("informe o valor desejavel que você está disposta a pagar"))
```

```python
filtered_dataset = dataset[(dataset['price'] >= valor_minimo) & (dataset['price'] <= valor_maximo)]

fig = px.scatter(filtered_dataset, x='bedrooms', y='price',
                 title='Preço das Casas vs. Número de Quartos',
                 labels={'bedrooms': 'Número de Quartos', 'price': 'Preço'},
                 color='price',
                 color_continuous_scale='Viridis',
                 hover_data={'price': True, 'bedrooms': True})

fig.show()
```
